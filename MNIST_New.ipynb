```python
X, y = mnist["data"], mnist["target"]
X.shape
```

```python
#easier to deal with number than string
import numpy as np
y = y.astype(np.uint8)
```

```python

X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
# print(np.shape(X_train))
# print(type(X_train))
# mysillyarray = np.array(X_train[0])
# print(np.shape(mysillyarray))
# # X_train = np.stack(X_train, mysillyarray)
# my2darray = np.array([mysillyarray])
# print(np.shape(my2darray))
# NewX_train = []
# for i in X_train:
#     NewX_train.append(i)
# NewX_train.append(mysillyarray)

# print(np.shape(np.array(NewX_train)))
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt



def shiftRight(mnistmatrixinput):
    mnistmatrix = mnistmatrixinput.copy()
    rows = []
    index = 0
    newRow = []
    for item in mnistmatrix:
        newRow.append(item)

        index = index + 1

        if(index % 28 == 0 or index == len(mnistmatrix)):
            rows.append(newRow)
            newRow = []



    shiftedRows = []
    for row in rows: 
    
        zeroval = [0]
        newShiftedRow = np.append(zeroval, row[0:len(row)-1])
        

        shiftedRows.append(newShiftedRow)

    

    return ((np.asarray(shiftedRows).flatten()))



def shiftLeft(mnistmatrixinput):
    mnistmatrix = mnistmatrixinput.copy()
    rows = []
    index = 0
    newRow = []
    for item in mnistmatrix:
        newRow.append(item)

        index = index + 1

        if(index % 28 == 0 or index == len(mnistmatrix)):
            rows.append(newRow)
            newRow = []



    shiftedRows = []
    for row in rows: 

        zeroval = [0]
        newShiftedRow = np.append(row[1:len(row)], zeroval)


        shiftedRows.append(newShiftedRow)


    return ((np.asarray(shiftedRows).flatten()))


def shiftDown(mnistmatrixinput):
    mnistmatrix = mnistmatrixinput.copy()
    zeroslen28 = np.zeros(28)
    return np.append(zeroslen28, mnistmatrix[0:len(mnistmatrix)-28])
def shiftUp(mnistmatrixinput):
    mnistmatrix = mnistmatrixinput.copy()
    zeroslen28 = np.zeros(28)
    return np.append(mnistmatrix[28:(len(mnistmatrix))], zeroslen28)





```

```python
import copy
index = 0
#append the shifted versions of each digit
#also making sure to append the correct (same) y_train value
#this is called data augmentation or training set expansion 
# print(X_train[0])




newX_train = []

newy_train = []

for x in X_train:
    newX_train.append(x.copy())
    newy_train.append(y_train[index])
    left = shiftLeft(x.copy())
    right = shiftRight(x.copy())
    up = shiftUp(x.copy())
    down = shiftDown(x.copy())
    
    newX_train.append(left)
    newy_train.append(y_train[index])
    newX_train.append(right)
    newy_train.append(y_train[index])
    newX_train.append(up)
    newy_train.append(y_train[index])
    newX_train.append(down)
    newy_train.append(y_train[index])
    index = index + 1
    if(index == 59999):
        break
    print(np.shape(newX_train), "index", index)

print(np.shape(newX_train))
```
